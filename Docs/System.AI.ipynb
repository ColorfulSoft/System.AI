```
using System;
using System.IO;

public static class Program
{
  public static void Main()
  {
    var img = new byte[512, 512, 3];
    var rand = new Random();
    for(int y = 0; y < 512; y++)
    {
      for(int x = 0; x < 512; x++)
      {
        unchecked
        {
          img[y, x, 0] = (byte)((2 * x - y) % 255);
          img[y, x, 1] = (byte)((x - 3 * y) % 255);
          img[y, x, 2] = (byte)((x + y) % 255);
        }
      }
    }
    imageio.imwrite("Mosaic.jpg", img);
  }
}
```
